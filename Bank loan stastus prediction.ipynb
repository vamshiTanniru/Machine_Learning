## a. Identify the Target Variable and Splitting the Data into Train and Test

````python
# Identify the inputs(X) and output (y)

y = df1['loan_status']
X = df1[['person_age', 'person_income', 'person_home_ownership',
         'person_emp_length', 
        'loan_intent', 'loan_grade',
         'loan_amnt','loan_int_rate','loan_percent_income',
        'cb_person_default_on_file','cb_person_cred_hist_length']]
````

````python
# Split into train and test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size = 0.75, random_state = 100)
````

````python
X_train.head()
````

````python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
````

## b. Seperate the Categorial and Numerical Columns:

````python
X_train.head()
````

````python
X_train.dtypes
````

````python
X_train_cat = X_train.select_dtypes(include=['object'])
X_train_cat.head()
````

````python
X_train_num=X_train.select_dtypes(include=['int64','float64'])
X_train_num
````

## c. Scaling the Numerical Features

````python
X_train_num.head()
````

````python
# scaling the numeroical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# column names are (annoyingly) lost after Scaling
# (i.e. the dataframe is converted to a numpy ndarray)

X_train_num_rescaled = pd.DataFrame(scaler.fit_transform(X_train_num),
                                   columns = X_train_num.columns,
                                   index = X_train_num.index)

X_train_num_rescaled.head()
````

````python
X_train_num.describe()
````

````python
print('Number of Numerical Feature:', scaler.n_features_in_)
print('Mean of each column:', scaler.mean_)
print('std of each column:', np.sqrt(scaler.var_))
````

## d. Applying OneHotEncoding on Categorical Columns

````python
X_train_cat['person_home_ownership'].value_counts(normalize= True)
````

````python
X_train_cat['loan_intent'].value_counts(normalize= True)
````

````python
X_train_cat['loan_grade'].value_counts(normalize= True)
````

````python
X_train_cat_ohe_df=X_train_cat.drop(['loan_grade'],axis=1)
````

````python
# OneHotEncoding the categorical features

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop= 'first', sparse= False)

# column names are (annoyingly) Lost after OneHotEncoding
# (i.e. the dataframe is converted to a numpy ndarray)\

X_train_cat_ohe = pd.DataFrame(encoder.fit_transform(X_train_cat_ohe_df),
                              columns = encoder.get_feature_names_out(X_train_cat_ohe_df.columns),
                              index = X_train_cat_ohe_df.index)

X_train_cat_ohe.head()
````

````python
encoder.categories_
````

## label Encoding

````python
X_train_cat.head()
````

````python
X_train_cat_le=pd.DataFrame(index=X_train_cat.index)
````

````python
X_train_cat_le.head()
````

````python
X_train_cat['loan_grade'].unique()
````

````python
values={'A':7,'B':6,'C':5,'D':4,'E':3,'F':2,'G':1}
X_train_cat_le['loan_grade']=X_train['loan_grade'].apply(lambda x: values[x])
X_train_cat_le.head()
````

## e.Concatinating the Encoded Categorical Features and Rescaled Numerical Features

````python
X_train_transformed = pd.concat([X_train_num_rescaled, X_train_cat_ohe,X_train_cat_le], axis=1)

X_train_transformed.head()
````

## logistic regression

````python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(X_train_transformed,y_train)


# prediction
y_test_pred = logistic.predict(X_test_transformed)

temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})

temp_df.head()

# Accuary
from sklearn import metrics

a = metrics.accuracy_score(y_test, y_test_pred)
print('Accuracy :',a)

sns.histplot(y_test, color='blue', alpha=0.5)
sns.histplot(y_test_pred, color='red', alpha=0.5);
````

## KNN classification

## Decision Tree Classsification

````python
from sklearn.tree import DecisionTreeClassifier
classification = DecisionTreeClassifier()
classification.fit(X_train_transformed, y_train)

y_test_pred = classification.predict(X_test_transformed)

temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})

temp_df.head()

sns.histplot(y_test, color='blue', alpha=0.5)
sns.histplot(y_test_pred, color='red', alpha=0.5);


from sklearn.ensemble import GradientBoostingClassifier
classifier = GradientBoostingClassifier()
classifier.fit(X_train_transformed, y_train)

y_test_pred = classifier.predict(X_test_transformed)

f = metrics.accuracy_score(y_test, y_test_pred)
print('Accuracy ;',f)
````
